```python
import sys
import time
import requests
import re
from bs4 import BeautifulSoup

webpage = requests.get("https://letterboxd.com/film/the-last-airbender/ratings/")
soup = BeautifulSoup(webpage.content, 'html.parser')
ratings = soup.find_all('section', class_="film-rating-group")
users = []
for rating in ratings:
    users_soup = rating.find_all('a', class_="avatar")
    i=0
    for user in users_soup:
        if i==20: break
        i += 1
        users.append(user['href'])
    break
```

```python
import sys
import time
import requests
import re
from bs4 import BeautifulSoup
#mlist = dict()
#users = ["/iaraya/"]
for u in users:  
    mlist[u] = dict()
    page = 1; validpage = True
    while validpage:
        webpage = requests.get("https://letterboxd.com"+u+"films/page/"+str(page)+"/")
        soup = BeautifulSoup(webpage.content, 'html.parser')
        validpage = False
        for span in soup.find_all('div', class_="really-lazy-load"):
            rating = span.parent.find_all('span', class_="rating");
            if len(rating)>0:
                x = re.search(r"\brated-(.*)\"", str(rating[0]))
                if x!=None:
                    validpage = True
                    mlist[u][(span['data-target-link'].split('/')[-2])]=int(x.groups()[0])
        time.sleep(1)
        page += 1
```
