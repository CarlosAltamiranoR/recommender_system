```python
import csv 
d2m = dict()

with open('ratings.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            line_count += 1
        else:
            director_list = row[12].split(', ')
            for d in director_list:
                if d in d2m:
                    d2m[d][1] = (d2m[d][1] * d2m[d][0] + float(row[1]))/(d2m[d][0]+1)
                    d2m[d][0] += 1
                else:
                    d2m[d] = [1,float(row[1])]
            
            line_count += 1
    print(f'Processed {line_count} lines.')
```
